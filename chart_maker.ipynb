```python
# importing pandas library
import pandas as pd
# import matplotlib library
import matplotlib.pyplot as plt

import numpy as np

```

```python
# Original Song vs. Conditional Generated Song

# creating dataframe
df = pd.DataFrame({
    'Name': ['MF Doom', 'Chance the Rapper', 'Eyedea', 'Drake'],
    'Original Song RD': [1.388 , 1.348, 1.244, 1.204] , 
    'Ghostwriter Avg. RD': [1.150, 0.921, 1.095, 0.900]
})
  
# plotting graph
ax = df.plot(x="Name", y=['Original Song RD', 'Ghostwriter Avg. RD'], kind="bar", title="Original Artist Song vs. Ghostwriter")
ax.set_xlabel("Author of Original Song")
ax.set_ylabel("Rhyme Density")
```

```python
# Original Song vs. Conditional Gerated Song

# creating dataframe
df = pd.DataFrame({
    'Name': ['Lupe Fiasco', 'Freddie Gibbs', 'Danny Brown', 'Aesop Rock', "Ghostwriter" , 'Lauryn Hill'],
    'RD': [1.067 , 1.064 , 1.043, 0.982, 0.958, 0.854]  
    
})
  
# plotting graph
ax = df.plot(x="Name", y=['RD'], kind="bar", title="Artist RD vs. Ghostwriter Unconditioned RD")
ax.set_xlabel("Original Artist")
ax.set_ylabel("Rhyme Density")
```

```python
mf_word = 266
gpt2_mfwordavg = np.mean(np.array([118, 136, 103, 91, 129]))

ctr_word = 210
gpt2_ctrwordavg = np.mean(np.array([117, 115, 117, 120, 131]))

eye_word = 225
gpt2_eyewordavg = np.mean(np.array([104, 122, 115, 98, 133]))

drake_word = 284
gpt2_drakewordavg = np.mean(np.array([133, 123, 140, 119, 160]))

# creating dataframe
df = pd.DataFrame({
    'Name': ['MF Doom', 'Chance the Rapper', 'Eyedea', 'Drake'],
    'Original Artist': [mf_word , ctr_word, eye_word, drake_word] , 
    'Ghostwriter': [gpt2_mfwordavg, gpt2_ctrwordavg, gpt2_eyewordavg, gpt2_drakewordavg]
})
  
# plotting graph
ax = df.plot(x="Name", y=['Original Artist', 'Ghostwriter'], kind="bar", title="Original Artist Song vs. Ghostwriter Unique Word Count")
ax.set_xlabel("Original Artist")
ax.set_ylabel("Unique Word Count")
```
